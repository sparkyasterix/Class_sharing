```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
from sklearn.datasets import make_classification
import seaborn as sns
from sklearn.linear_model import LinearRegression

data = {
        'GPA': [3.0, 3.7, 3.3, 2.5, 4, 3.2, 3.9, 3.0],
        'Hours Studied': [2, 8, 3, 1, 12, 7, 9, 5]
        }

df = pd.DataFrame(data, columns = ['GPA', 'Hours Studied'])


# Using the matplotlib library plotting the data
X = (df['Hours Studied'].values).reshape(-1,1)
Y = df['GPA'].values


model = LinearRegression().fit(X, Y)
r_sq = model.score(X, Y)
print('coefficient of determination R2:', r_sq)
print('intercept: B0', model.intercept_)
print('slope:B1', model.coef_)

Y_pred = model.predict(X)
print('predicted response:', Y_pred, sep='\n')


# Plot the data point
plt.scatter(X, Y, color='#ff0000', label='Data Point')
plt.scatter(X, Y_pred, color='#999999', label='Predicted Point')
plt.plot(X, Y_pred, color='#777777', label='Linear Regression')

#axis labels
plt.xlabel('Hours Studied')
plt.ylabel('GPA')

plt.legend()
plt.show()


```
